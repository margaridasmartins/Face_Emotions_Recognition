## Score table

```python
import matplotlib.pyplot as plt

columns = rows = [0.001, 0.01, 0.1, 1, 10, 100]

values = [
    [.25, .27, .24, .24, .25, .20],
    [.25, .27, .24, .24, .25, .20],
    [.27, .30, .24, .24, .25, .20],
    [.35, .39, .26, .20, .20, .20],
    [.37, .40, .27, .20, .20, .20],
    [.36, .39, .27, .20, .20, .20]
]

# Colorize cells depending on value
vals = [o for items in values for o in items]
normal = plt.Normalize(min(vals), max(vals))
cellColours=plt.cm.Blues(normal(values))

# Draw figure
plt.figure(figsize=(15,15))

the_table = plt.table(
  cellText=[[f"{o:.2f}" for o in items] for items in values],
  rowLabels=rows,
  colLabels=columns,
  loc='center',
  cellColours=cellColours
)
the_table.scale(1.2, 5)
the_table.set_fontsize(20)
plt.axis('off')
plt.axis('tight')
# plt.subplots_adjust(left=0.2, top=10)
plt.tight_layout()

plt.savefig(f"./metrics/score.png")
```
